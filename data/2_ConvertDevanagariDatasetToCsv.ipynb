**How to use:**
1. Create ./hindi directory
2. Copy the "Train" and "Test" folders from the DevanagariHandwrittenCharacterDataset.zip file to above directory
3. Run all cells in order

```python
import skimage.io as io
import os
```

```python
# Rename directories
os.rename("./hindi/Train", "./hindi/train")
os.rename("./hindi/Test", "./hindi/test")
```

```python
# Define path to training and testing directories
train_dir = "./hindi/train"
test_dir = "./hindi/test"
```

```python
# Rename subdirectories to match class labels
def rename_subdirs(dir_path):
    filenames = os.listdir(dir_path)
    for filename in filenames:
        if filename.startswith("character"):
            new_name = str(int(filename.split("_")[1]) + 26)
        else:
            new_name = str(int(filename.split("_")[1]) + 26 + 36 + 1)
        os.rename(os.path.join(dir_path, filename), os.path.join(dir_path, new_name))

rename_subdirs(train_dir)
rename_subdirs(test_dir)
```

```python
def create_csv(output_csv_path, images_dir_path):
    # Get a file handle to output csv file opened in write mode
    f = open(output_csv_path, "w")

    # Create csv header row
    f.write("Label")
    for i in range(1024):
        f.write(",{}".format(i))
    f.write("\n")
    
    # Get list of subdirectories
    subdirs = os.listdir(images_dir_path)
    for subdir in subdirs:
        # Label for current image is the name of subdirectory under which it is present
        cur_label = subdir
        # Save path to current subdirectory in a variable
        subdir_path = os.path.join(images_dir_path, subdir)
        # Get filenames of images present under subdirectory
        filenames = os.listdir(subdir_path)
        
        # Loop over each image
        for filename in filenames:
            # Get the full path to the image file so it can be read
            img_path = os.path.join(subdir_path, filename)
            
            # Read the image data into memory
            img = io.imread(img_path)
            
            # Write the data into the csv file
            f.write("{}".format(cur_label))
            for j in img.reshape(32 * 32):
                f.write(",{}".format(j))
            f.write("\n")
            
            print("Processed {}".format(filename))

    # Close the file handle to save changes
    f.close()
```

```python
# Create csv for training dataset
create_csv("./hindi/train.csv", train_dir)
```

```python
# Create csv for testing dataset
create_csv("./hindi/test.csv", test_dir)
```

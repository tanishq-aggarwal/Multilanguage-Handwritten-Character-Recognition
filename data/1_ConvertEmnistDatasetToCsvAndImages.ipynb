**How to use:**
1. Create ./english directory
2. Copy EMNIST letters dataset (in .gz format) to above directory
3. Run all cells in order

```python
from mnist import MNIST
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
# Create MNIST class object with path to directory containing dataset ("./english")
mndata = MNIST("./english")
mndata.select_emnist("letters")
```

```python
def write_csv_and_images(df, output_csv_path, output_images_path):
    # File handle to csv file opened in write mode
    f = open(output_csv_path, "w")

    # Create csv header row
    f.write("Label")
    for i in range(1024):
        f.write(",{}".format(i))
    f.write("\n")

    # Loop over each row in dataframe containing images data
    for i in range(len(df)):
        # Grab label associated with current row
        cur_label = labels[i]
        # Convert 28x28 image to 32x32 image by surrounding with black pixels
        img = pd.DataFrame(df.iloc[i].values.reshape(28, 28))
        img.insert(0, -1, 0)
        img.insert(0, -2, 0)
        img.insert(30, 28, 0)
        img.insert(31, 29, 0)
        img.columns = np.arange(32)
        img.loc[-1] = np.zeros(32, dtype=int)
        img.loc[-2] = np.zeros(32, dtype=int)
        img.loc[28] = np.zeros(32, dtype=int)
        img.loc[29] = np.zeros(32, dtype=int)
        img.index = img.index + 2
        img.sort_index(inplace=True)
        
        # Write image data to output csv file
        f.write("{}".format(cur_label))
        for j in img.values.reshape(32 * 32):
            f.write(",{}".format(j))
        f.write("\n")
        
        # Write image to output directory
        if not os.path.isdir(os.path.join(output_images_path, str(cur_label))):
            os.makedirs(os.path.join(output_images_path, str(cur_label)))
        plt.imsave(os.path.join(output_images_path, str(cur_label), "{}.png".format(i)), img)

        print("Processed #{}".format(i))
    
    # Close handle to save changes
    f.close()
```

```python
# Load training dataset into memory
images, labels = mndata.load_training()
# Convert images' pixels into pandas dataframe
df = pd.DataFrame(images)
# Delete "images" variable from memory to save space
del images
```

```python
# Create csv and images for training dataset
write_csv_and_images(df, "./english/train.csv", "./english/train")
```

```python
# Load testing dataset into memory
images, labels = mndata.load_testing()
# Convert images' pixels into pandas dataframe
df = pd.DataFrame(images)
# Delete "images" variable from memory to save space
del images
```

```python
# Create csv and images for testing dataset
write_csv_and_images(df, "./english/test.csv", "./english/test")
```
